```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Given data points
time = np.array([0, 1, 2, 3, 4, 5])
temperature = np.array([20, 25, 30, 35, 40, 45])

# Define a model function (e.g., linear, quadratic, exponential)
def model_function(t, a, b):
    return a * t + b  # Linear model

# Fit the curve to the data
params, _ = curve_fit(model_function, time, temperature)

# Generate predicted values using the fitted parameters
predicted_temperature = model_function(time, *params)

a, b = params
print("Estimated parameters:")
print("Slope (a):", a)
print("Intercept (b):", b)

# Plot the original data points and the fitted curve
plt.scatter(time, temperature, label='Data Points')
plt.plot(time, predicted_temperature, 'r-', label='Fitted Curve')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Curve Fitting Example')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Given data points
time = np.array([0, 1, 2, 3, 4, 5])
Covid19 = np.array([100, 200, 400, 800, 1600, 3200])

# Define an exponential model function
def exponential_model(t, a, b):
    return a * np.exp(b * t)

# Fit the curve to the data
params, _ = curve_fit(exponential_model, time, Covid19)

# Generate predicted values using the fitted parameters
predicted_Covid19 = exponential_model(time, *params)

a, b = params
print("Estimated parameters:")
print("Slope (a):", a)
print("Intercept (b):", b)

# Plot the original data points and the fitted curve
plt.scatter(time, Covid19, label='Data Points')
plt.plot(time, predicted_Covid19, 'r-', label='Fitted Curve')
plt.xlabel('Time')
plt.ylabel('Covid19')
plt.title('Exponential Curve Fitting Example')
plt.legend()
plt.grid(True)
plt.show()
```
