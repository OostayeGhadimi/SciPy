### Polynomial Interpolation Example

Suppose we have the following set of data points:

\[
\begin{array}{|c|c|}
\hline
x & y \\
\hline
0 & 1 \\
1 & 2 \\
2 & 3 \\
3 & 4 \\
\hline
\end{array}
\]

We'll use polynomial interpolation to find a polynomial function that passes through these data points.



```python
import numpy as np
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt

# Given data points
x_data = [0, 1, 2, 3]
y_data = [1, 2, 3, 4]

# Perform polynomial interpolation
poly_interpolation = lagrange(x_data, y_data)

# Evaluate the polynomial at a specific point
x_interpolate = 2.5
y_interpolate = poly_interpolation(x_interpolate)

print(f"Interpolated value at x = {x_interpolate}: y = {y_interpolate}")


# Generate points for plotting the interpolated polynomial
x_interpolate = np.linspace(min(x_data), max(x_data), 100)
y_interpolate = poly_interpolation(x_interpolate)

# Plot the original data points
plt.plot(x_data, y_data, 'bo', label='Data Points')

# Plot the interpolated polynomial
plt.plot(x_interpolate, y_interpolate, 'r-', label='Interpolated Polynomial')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Interpolation')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
